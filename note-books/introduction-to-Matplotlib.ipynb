## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = np.array([1,2,3,4,5,6])
y = np.array([10,13,12,11,14,15])

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (save the whole figure)
fig.savefig('./images/sample-plot.png')
```

## SubPlots
### Two option for subPlots

```python
# 1st Option
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(car_p_num.keys(),car_p_num.values())
ax4.hist(np.random.randn(1000));
```

```python
# 2nd Option 
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# make plots by indexes
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(car_p_num.keys(),car_p_num.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one should we use? (pyplot or matplotlib OO method?)
* when we plotting something quickly, pyplot is good to use.
* when we plotting something more advanced, OO method is more useful and better.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
## pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(10,6));
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);
# customize the plot
ax.set(title='Heart Disease and Cholestrol level',
       xlabel='Age',
       ylabel='Cholestrol');
# Add a legend
ax.legend(*scatter.legend_elements(),title='Target');
# add average line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True) # this line allow us to use one label for x axis because they are same.
# add ax0 data
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])
# customize ax0
ax0.set(title='Heart disease based on Age and Cholestrol level',
        ylabel='Cholestrol');
# add legend
ax0.legend(*scatter.legend_elements(),title='Target');
# add meanline
ax0.axhline(over_50['chol'].mean(), linestyle='--')

# add ax1 data
scatter = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target']);
# customize ax1
ax1.set(title='Heart disease based on Age and Heart rate',
        xlabel='Age',
        ylabel='Max Heart Rate');
# add Legend
ax1.legend(*scatter.legend_elements(), title='Target')
# add meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--')

# add title for figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```

```python
# change the style
plt.style.use("seaborn-whitegrid")
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter'); # change the color scheme
# set x limitation
ax.set_xlim([50,80])
# customize the plot
ax.set(title='Heart Disease and Cholestrol level',
       xlabel='Age',
       ylabel='Cholestrol');
# Add a legend
ax.legend(*scatter.legend_elements(),title='Target');
# add average line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
# Customising x and y axis limitation
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True) # this line allow us to use one label for x axis because they are same.
# add ax0 data
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter') # change color scheme
# change x limitation
# ax0.set_xlim([50,80]) # because they share their axises it's not necessary to chagne ax0 x limitation.

# customize ax0
ax0.set(title='Heart disease based on Age and Cholestrol level',
        ylabel='Cholestrol');
# add legend
ax0.legend(*scatter.legend_elements(),title='Target');
# add meanline
ax0.axhline(over_50['chol'].mean(), linestyle='--', c='red') # c -> change line color

# add ax1 data
scatter = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'],
                      cmap='winter'); # change color scheme
# change x and y limitation
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# customize ax1
ax1.set(title='Heart disease based on Age and Heart rate',
        xlabel='Age',
        ylabel='Max Heart Rate');
# add Legend
ax1.legend(*scatter.legend_elements(), title='Target')
# add meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--',color='red') # change line color, c

# add title for figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```
